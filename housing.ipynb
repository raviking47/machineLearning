```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics 
```

```python
datasetk = sklearn.datasets.load_boston()
```

```python
print(datasetk)
```

```python
dataset = pd.DataFrame(datasetk.data,columns=datasetk.feature_names)
```

```python
dataset.head(5)
```

```python
# add the Target Column in our dataset 
dataset['price'] = datasetk.target
```

```python
X = dataset.drop(['price'],axis=1)
Y = dataset['price']
```

```python
print(X)
print(Y)
```

```python
X_train,X_test,Y_train,Y_test =  train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_test.shape,X_train.shape)
```

```python
model = XGBRegressor()
model.fit(X_train,Y_train)
```

```python
print(model)
```

```python
predication  = model.predict(X_train)
print(predication)
```

```python
score_1  = metrics.r2_score(Y_train,predication)

score_2 = metrics.mean_absolute_error(Y_train,predication)
```

```python
print(score_1,score_2)
```

```python
predication_test = model.predict(X_test)
score_1= metrics.r2_score(Y_test,predication_test)
score_2  = metrics.mean_absolute_error(Y_test,predication_test)
```

```python
print(score_1,score_2)
```
